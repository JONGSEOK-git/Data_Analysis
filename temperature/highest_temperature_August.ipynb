```python
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]
```

```python
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08' :
            result.append(float(row[-1]))
plt.plot(result,'hotpink')
plt.show()
```

```python
print(len(result))
```

```python
plt.plot(result, 'r')
plt.figure(figsize = (10, 2))
plt.show()
```
