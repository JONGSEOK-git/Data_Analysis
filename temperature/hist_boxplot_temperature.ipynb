```python
f=open('seoul.csv')
data=csv.reader(f)
next(data)
jan = []
aug = []
```

```python
for row in data :
    month = int(row[0].split('-')[1])
    if row[-1] != '' :
        if month == 8 :
            aug.append(float(row[-1]))
        if month == 1 :
            jan.append(float(row[-1]))
plt.hist(jan, bins=100, color='r', label='january')
plt.hist(aug, bins=100, color='b', label='august')
plt.legend()
plt.show()
```

```python
import numpy as np
result=[]
for i in range(13):
    result.append(random.randint(1,1000))
print(sorted(result))
print('1/4: ' +str(np.percentile(result,25))) # 위치에 있는 값들을 알려주는 코드
print('2/4: ' +str(np.percentile(result,50)))
print('3/4: ' +str(np.percentile(result,75)))
plt.boxplot(result)
plt.show()
```

```python
f2=open('seoul.csv')
data2=csv.reader(f2)
next(data2)
high=-999

a_month=[[],[],[],[],[],[],[],[],[],[],[],[]]
for row in data2:
    if row[-1] != '' :
        a_month[int(row[0].split('-')[1])-1].append(float(row[-1]))
        
for i in a_month[7] :
    if high < i :
        high = i

print(high)
print(str(np.percentile(a_month[7], 100)))
        
plt.boxplot(a_month)
plt.show()                                         
```
