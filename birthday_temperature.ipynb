```python
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
```

```python
high = []
low = []
for row in data:
    if row[-1] != '' and row[-2] != '' :
        if row[0].split('-')[1] == '01' and row[0].split('-')[2] == '01' :
            high.append(float(row[-1]))
            low.append(float(row[-2]))        
plt.title("birthday_temperatrue")
plt.plot(high, color="blue", label='HIGH')
plt.plot(low, color = 'red', label='LOW')
plt.legend()
plt.show()
```
