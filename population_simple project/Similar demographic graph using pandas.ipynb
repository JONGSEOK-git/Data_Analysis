```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('age.csv', encoding='cp949', index_col=0)
df.head()
```

```python
df = df.div(df['2019년02월_계_총인구수'], axis=0)
del df['2019년02월_계_총인구수'], df['2019년02월_계_연령구간인구수']
```

```python
name=input('원하는 지역의 이름을 입력해주세요 : ')
a= df.index.str.contains(name) # 인덱스 문자열에 원하는 문자열이 포함된 행을 찾아냄
df2 = df[a]
df2
```

```python
plt.rc('font', family='Malgun Gothic')
df2.T.plot()
plt.show()
```

```python
x=df.sub(df2.iloc[0], axis=1) #df2==>선택한 지역이 있다.
y=np.power(x,2) # x**2
z=y.sum(axis=1)
```

```python
i = z.sort_values().index[:5]
```
