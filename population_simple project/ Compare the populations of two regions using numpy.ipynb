```python
import csv
import numpy as np

f=open('age.csv')
data=csv.reader(f)

next(data)

home=[]
home2=[]

name = input('1. 인구 구조가 알고싶은 지역의 이름(읍면동 단위)을 입력해주세요 :')
name2 = input('2. 인구 구조가 알고싶은 지역의 이름(읍면동 단위)을 입력해주세요 :')

for row in data:
    if name in row[0]:
        home=np.array(row[3:], dtype=int) / int(row[2].replace(',',''))
    if name2 in row[0]:
        home2=np.array(row[3:], dtype=int) / int(row[2].replace(',',''))
print(home)
print(home2)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.title(name+' 지역의 인구 구조')
plt.plot(home, label = "신도림")
plt.plot(home2, label = "화촌면")
plt.legend()
plt.show()
```
