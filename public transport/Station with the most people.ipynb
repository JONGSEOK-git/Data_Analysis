```python
import csv
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

result=[]
for row in data:
    row[4:]=map(int, row[4:])
    result.append(row[10])

print(len(result))
print(result)
```

```python
import matplotlib.pyplot as plt
result.sort() # 오름차순
plt.bar(range(len(result)), result)
plt.show()
```

```python
f2=open('subwaytime.csv')
data2=csv.reader(f2)

next(data2)
next(data2)
result2=[]

max = 0 # 승차
max2 = 0 # 하차
max_station = ''
max2_station = ''

for row in data2:
    row[4:]=map(int, row[4:])
    result2.append(sum(row[10:15:2]))
    if(sum(row[10:15:2])> max):
        max=sum(row[10:15:2])
        max_station=row[1] +' '+ row[3]
    a=row[11:16:2]
    if sum(a)>max2:
        max2=sum(a)
        max2_station=row[1]+' '+ row[3]
print(max_station, max)
print(max2_station, max2)
```

```python
result2.sort()
plt.bar(range(len(result2)), result2)
plt.show()
```

```python
f3=open('subwaytime.csv')
data3=csv.reader(f3)

next(data3)
next(data3)

max=0
max_station=''

time=input('시간을 입력해주세요 : ')
result_time = 2*(int(time)) - 4

for row in data3 :
    row[result_time]=int(row[result_time])
    if row[result_time] > max :
        max = row[result_time]
        max_station = row[1] + ' ' + row[3]
print(max_station, max)
```
