```python
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data=csv.reader(f)

next(data)
next(data)

max=[0]*24
max_station = ['']*24

for row in data:
    row[4:]=map(int, row[4:])
    for i in range(24):
        j = i*2+4
        if row[j] > max[i] :
            max[i]=row[j]
            if i+4 <=24:
                max_station[i] = row[1] + ' ' + row[3]+'('+str(i+4)+')'
            else :
                max_station[i] = row[1] + ' ' + row[3]+'('+str(i-21)+')'
print(max_station, max)
```

```python
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), max)
plt.xticks(range(24), max_station, rotation=90)
plt.show()
```

```python
f2=open('subwaytime.csv')
data2=csv.reader(f2)

next(data2)
next(data2)

mx=[0]*24
mx_station=['']*24

for row in data2:
    row[4:]=map(int, row[4:])
    for j in range(24):
        b= row[5 + j * 2]
        if b> mx[j] :
            mx[j]=b
            mx_station[j]=row[3] + '(' +str(j+4)+ ')'
            
```

```python
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx, color='b')
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
f3=open('subwaytime.csv')
data3=csv.reader(f3)

next(data3)
next(data3)

s_in=[0]*24
s_out=[0]*24
station=[]

for row in data3:
    row[4:]=map(int, row[4:])
    for i in range(24):
        s_in[i] += row[i*2+4] 
        s_out[i] += row[i*2+5]
    name = row[1] + row[3]
    station.append(name)
```

```python
plt.rc('font', family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24), range(4,28))
plt.show()
```
