```python
import csv
import matplotlib.pyplot as plt

f=open('age.csv')
data=csv.reader(f)
```

```python
result = []
for row in data:
    if '신도림' in row[0]:
        for i in row[3:]:
            result.append(int(i))
print(result)
```

```python
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
d=open('age.csv')
data2=csv.reader(d)
result2=[]
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')

for row in data2:
    if name in row[0]:
        for i in row[3:] :
            result2.append(int(i.replace(',','')))
        
plt.rc('font', family='Malgun Gothic')
plt.style.use('ggplot')
plt.title(name + '지역의 인구 구조')
plt.plot(result2)
plt.show()
```
