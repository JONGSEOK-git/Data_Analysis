```python
import csv
f=open('age.csv')
data=csv.reader(f)

result= []
for row in data :
    if '신도림' in row[0] :
        for i in row[3:] :
            result.append(int(i.replace(',','')))
            
plt.bar(range(101), result)
plt.show()
```

```python
plt.barh(range(101), result)
```
