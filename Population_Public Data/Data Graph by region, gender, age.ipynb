```python
import csv
f=open('gender.csv')
data=csv.reader(f)
```

```python
m=[]
f=[]

name=input('동네를 입력해주세요 : ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break 
```

```python
import matplotlib.pyplot as plt
plt.plot(m, label='Male')
plt.plot(f, label='Female')
plt.legend()
plt.show()
```
